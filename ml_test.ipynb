model selection - train / test split 

```python
x = iris.drop(['species'], axis=1).copy()
y = iris['species'].copy()
print (x.shape)
print(y.shape)
set(list(y))
```

```python
#encoding the data for 'species' column

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=5)
```

```python
y_test
```

model creation

```python
#model creation - K nearest neighbors classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

k_range = list(range(1,25))
scores_train=[]
scores_test=[]

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred_train= knn.predict(x_train)
    y_pred_test=knn.predict(x_test)

    scores_train.append(metrics.accuracy_score(y_train, y_pred_train))
    scores_test.append(metrics.accuracy_score(y_test, y_pred_test))
```

```python
df_plot=pd.DataFrame({'scores_train':scores_train, 'scores_test':scores_test})
print("confusion matrix of train sets")
print(confusion_matrix(y_train, y_pred_train))
print("confusion matrix of train sets")
print(confusion_matrix(y_test, y_pred_test))
```

```python
df_plot
```

```python
df_plot.plot(figsize=(12,10), title = 'accurancy of knn classifier model fitting on train and test sets')
```

user elbow method to find the opimal number of clusters

```python
from sklearn.cluster import KMeans
cs = []

for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    cs.append(kmeans.inertia_)
plt.plot(range(1,11), cs)
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

#deep learning concept

#multilayer perceptron 
#forward propogation
#cost function - input - deep learning - output (predict output as close as possible). c = 1/m E(y-a)2
#gradient descend - optimization algo to minimize cost 
#learning rate - is the amount of minimization in cost function, rate at which model learn , graph is optimal 
#backpropogation - feedback mechanism, such that error is reduced in future layer. 
#batches - entire input in 1 go, we divide input into batches, training in batches makes model more generalised. 
#epochs - what is epochs, single training iteration of all the batches, both in forward and backword propogation
#dropout, technic, to avoid overfitting, don't consider output of multiple network, and then take output. 
#batch normalization - reduce, the number of iterations, no of epoch that gets generated, weight need to be adjusted so that subsequent cycle has better accuracy , we normalize the data before sending to the next layer
